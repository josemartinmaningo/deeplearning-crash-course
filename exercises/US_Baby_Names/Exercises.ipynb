### Step 6. Is there more male or female names in the dataset?

```python
count = baby_names.groupby('Gender').sum()
count = count.drop(['Unnamed: 0', 'Id', 'Year'], axis = 1)
F_count = count.iloc[0, 0]
M_count = count.iloc[1, 0]
print("Female = ", F_count)
print("Male = ", M_count)
if F_count > M_count:
    print("There are more females than males!")
elif F_count < M_count:
    print("There are more males than females!")
else:
    print("The number of females and males are the same!")
```
